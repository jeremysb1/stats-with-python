```python
import numpy as np
```

```python
collie_weights = np.array([51.2,53.1,53.1,53.9,59.0,
   59.3,60.2,61.1,67.2,67.6])
```

```python
collie_weight_mean = np.mean(collie_weights)
```

```python
print("Collie Weight Mean (lbs):", collie_weight_mean)
```

```python
# Three exams and a final score
exam_scores = [85, 90, 88, 92]
```

```python
# Weighting for each exam
weights = [.2, .2, .2, .4]
```

```python
# Calculate weighted average
weighted_average = sum(e*w for e, w in zip(exam_scores, weights)) / sum(weights)
```

```python
print("Final grade: ", weighted_average)
```

```python
print("Using NumPy: ", np.average(exam_scores, weights=weights))
```

```python
bar_tabs = [7, 18, 19, 20, 21, 22, 25, 120, 1500]
```

```python
mean = np.mean(bar_tabs)
median = np.median(bar_tabs)
```

```python
print("Mean: ", mean)
print("Median: ", median)
```
