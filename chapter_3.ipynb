```python
from random import random

heads, tails = 0, 0
p = 0.5

for i in range(1000):
    if random() < p:
        heads += 1
    else:
        tails += 1

print(f"HEADS: {heads}, TAILS: {tails}")
```

```python
import matplotlib.pyplot as plt

p = 0.7
x = [0, 1]
probabilities = [1-p, p]
```

```python
# creating the plot

plt.figure(figsize=(10, 6))
bars = plt.bar(x, probabilities,
               color=['#002d8b'],
               hatch=['///', 'xxx'],
               edgecolor='black')

plt.title('Bernouli Distribution: MacBook Water Damage Recovery', pad=20)
plt.xlabel('Outcome (0 = Not Recovered, 1 = Recovered)')
plt.ylabel('Probability')
```

```python
for i, prob in enumerate(probabilities):
    plt.text(i, prob, f'{prob: 2f}', ha='center', va='bottom')
```

```python
plt.xticks(x, ['Not Recoverec\n(0)', 'Recovered\n(1)'])
```
